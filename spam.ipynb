```python
from flask import Flask, request, jsonify
import joblib
import numpy as np
from sentence_transformers import SentenceTransformer

app = Flask(__name__)

model = joblib.load('logreg_model.pkl')
model1 = SentenceTransformer('all-MiniLM-L6-v2')

@app.route('/predict', methods=['POST'])
def predict():
    data = request.get_json()
    sentence = data['text']
    X_emb = model1.encode(sentence)
    pred = model.predict(np.array(X_emb).reshape(1, -1))
    return jsonify({'prediction': int(pred[0])})
```

```python
import pandas as pd
df = pd.read_csv(r"spam (Text).csv")

X=df['Message']
Y=df['Category']

from sklearn.feature_extraction.text import CountVectorizer
X_CV = CountVectorizer().fit_transform(X)

from sklearn.model_selection import train_test_split
X_train, X_test, Y_train, Y_test = train_test_split(X_CV, Y)
```

```python
print(X_CV)
```

```python

from sklearn.linear_model import LogisticRegression
from sklearn.metrics import confusion_matrix

model = LogisticRegression()
model.fit(X_train, Y_train)
pred = model.predict(X_test)
cm=confusion_matrix(Y_test, pred)
print(cm)
acc=(cm[0,0]+cm[1,1])/sum(sum(cm))
print('Accuracy without TFIDF is ', acc)
```
